```python
import pandas as pd
import numpy as np

test = pd.read_csv('test.csv')
```

```python
test.Date = pd.to_datetime(test.Date)
```

```python
weather_clean = pd.read_csv('weather_dt_clean.csv')

weather_clean['Date'] = pd.to_datetime(weather_clean['Date'])
weather_clean.PrecipTotal.replace(regex=True,inplace=True,to_replace=r'T',value=0.005)
weather_clean.PrecipTotal = weather_clean.PrecipTotal.astype(float)

weather_cols = ['Tmax','Tmin','Tavg','DewPoint',
                'Heat','Cool','PrecipTotal',
                'StnPressure','ResultDir','AvgSpeed']

#List of lists that will have all the 15 days prior weather data
new_cols = [weather_cols]
#iterate through every row
for i in test.index:
    new_row = [] #create a new, empty row
    #create a new datafram containing all rows that occurred 15 days before the ith row
    #and were at the same station
    fif_days_before = weather_clean[weather_clean['Date'] == (test.loc[i].Date - pd.Timedelta('15 days'))]
    #print df.loc[i].Date
    #print len(fif_days_before)
    
    for j in weather_cols:
        new_row.append(np.mean([fif_days_before.iloc[0][j], fif_days_before.iloc[1][j]]))
        
            
    new_cols.append(new_row)
	
new_col_names = [i + "-15d" for i in weather_cols]
new_cols_df = pd.DataFrame(new_cols[1:],columns=new_col_names)
dums = pd.get_dummies(test.Trap,drop_first=True)
result = pd.concat([test, new_cols_df,dums], axis=1)
```
